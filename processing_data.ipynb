```python
import numpy as np

# 1. Load the file with allow_pickle=True and convert it back to a standard Python list/tuple
finance_data = np.load('models/processed_data/finance_data.npy', allow_pickle=True).tolist()

# 2. Unpack the three dictionaries stored inside
ticker_name_map, ticker_info_map, ticker_name_set_map = finance_data

# 3. Inspect the data! Let's print the first 5 items in the ticker map
print("First 5 Tickers:")
print(list(ticker_name_map.items())[:5])

# Let's see how many total finance companies your RAG system knows about:
print(f"\nTotal companies in database: {len(ticker_name_map)}")
```

```python
import numpy as np

# 1. Load the Grammy data
data = np.load('models/processed_data/grammy.npy', allow_pickle=True).tolist()

# 2. Check the top-level data type
print(f"Top-level Data Type: {type(data)}")

# 3. Safely inspect based on what it is
if isinstance(data, tuple) or isinstance(data, list):
    print(f"It contains {len(data)} items.")
    print("-" * 30)
    
    # Loop through and print a summary of each item
    for i, item in enumerate(data):
        print(f"\nItem {i+1} Type: {type(item)}")
        if isinstance(item, dict):
            print(f"Number of entries: {len(item)}")
            print(f"Sample Keys: {list(item.keys())[:3]}")
            
elif isinstance(data, dict):
    print(f"Number of entries in dictionary: {len(data)}")
    print(f"Sample Keys: {list(data.keys())[:5]}")
    
else:
    print(f"Data is a single object: {data[:100]}...")
```
